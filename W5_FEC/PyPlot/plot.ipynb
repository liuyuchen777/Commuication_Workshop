```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import special
# x label
x = np.linspace(0, 10, num=11)
# read data
y2 = np.loadtxt('awgn.txt')
y2 = y2[:,1]
# therotical data
y1 = 0.5*special.erfc(np.sqrt(np.exp(x*np.log(10)/10)))
# plot figure
plt.xlabel('$E_b/N_0 (dB)$')
plt.ylabel('Average BER')

plt.xlim(0,10)
plt.ylim(0.0000001, 1)

plt.semilogy(x, y1, label = "AWGN Theory")
plt.semilogy(x, y2, linewidth = 0,  marker = '^', markersize = 5, label = "AWGN Simulation")

plt.grid(linestyle='-.')

plt.legend(loc = 'upper right')

plt.show()

plt.savefig('awgn.eps', format = 'eps')
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import special
# x label
x = np.linspace(0, 30, num=31)
# read data
y2 = np.loadtxt('rayleigh.txt')
y2 = y2[:,1]
# therotical data
y1 = 0.5*(1-np.sqrt((10**(x/10))/(10**(x/10)+1)))
# plot figure
plt.xlabel('$E_b/N_0 (dB)$')
plt.ylabel('Average BER')

plt.xlim(0,30)
plt.ylim(0.00001, 1)

plt.semilogy(x, y1, label = "Rayleigh FECC Theory")
plt.semilogy(x, y2, marker = 'o', markersize = 5, label = "Rayleigh FECC Simulation")

plt.grid(linestyle='-.')

plt.legend(loc = 'upper right')

plt.show()

plt.savefig('rayleigh.eps', format = 'eps')
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0.0, 0.25, num=250)
y = np.sin(2 * np.pi * 50 * x) + np.sin(2 * np.pi * 120 * x)
y = y + np.random.randn(y.size)/2
plt.plot(x, y)
plt.title('Noisy Time Domain Signal')
plt.show()
```

```python
fft_y = np.fft.fft(y, n=251)
amy = fft_y * fft_y.conj() / 251
f = 1000 / 251 * np.arange(127)
plt.plot(f, amy[1:128])
plt.xlabel('Frequency (Hz)')
plt.title('Power Spectral Density')
plt.show()
```
